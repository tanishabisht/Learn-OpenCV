# Histograms
Graph/plot to get an overall idea of the intensity distribution of an image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = np.zeros((200,200), np.uint8)
img = cv2.rectangle(img, (0,100), (200,200), (255), -1)
img = cv2.rectangle(img, (0,50), (100,100), (127), -1)

img2 = cv2.imread('./Pictures/lena.jpg')
b,g,r = cv2.split(img2)
```

```python
# plt.hist(src.ravel(), maxPixelVal, pixelValRange)

plt.hist(img.ravel(), 256, [0,256])
plt.hist(img2.ravel(), 256, [0,256])

plt.hist(b.ravel(), 256, [0,256])
plt.hist(g.ravel(), 256, [0,256])
plt.hist(r.ravel(), 256, [0,256])

plt.show()
```

```python
# cv2.calcHist([images], [indexOfChannels], mask, histSize, range, hist=None, accumulate=None)

hist = cv2.calcHist([img2], [0], None, [256], [0,256])
plt.plot(hist)
```
