```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./Pictures/lena.jpg', -1)

# OpenCV reads images in BGR format
# Matplotlib reads images in RBG format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# To show many images in one window in matplotlib
titles = ['img1', 'img2', 'img3', 'img4']
images = [img, img, img, img]

for i in range(4):
    
    # create a subplot
    # plt.subplot(rows, col, imgNo)
    plt.subplot(2, 2, i+1)
    
    # show the images through matplotlib
    plt.imshow(images[i], 'gray')
    
    # display the titles of the images
    plt.title(titles[i])    
    
    # to hide the ticks in x and y axis
    plt.xticks([])
    plt.yticks([])
    
plt.show()
```
