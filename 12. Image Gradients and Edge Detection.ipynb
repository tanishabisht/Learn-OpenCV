# Image gradient
is a directional change in intensity or color in an image

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./Pictures/sudoku.png', cv2.IMREAD_GRAYSCALE)
```

## Laplacian method
returns negative values too which needs to be converted to unsigned 8 bit interger as its suitable for our output.

```python
laplacian = cv2.Laplacian(img, cv2.CV_64F, ksize=3)
laplacian = np.uint8(np.absolute(laplacian))
```

## Sobel method
returns negative values too which needs to be converted to unsigned 8 bit interger as its suitable for our output.

```python
sobelX = cv2.Sobel(img, cv2.CV_64F, 1, 0)
sobelX = np.uint8(np.absolute(sobelX))

sobelY = cv2.Sobel(img, cv2.CV_64F, 0, 1)
sobelY = np.uint8(np.absolute(sobelY))

sobelXY = cv2.Sobel(img, cv2.CV_64F, 1, 1)
sobelXY = np.uint8(np.absolute(sobelY))

sobelXorY = cv2.bitwise_or(sobelX, sobelY)
```

## Canny Edge detector
is an edge detection operator that uses a multi-stage algorithm to detect a wide range of edges in images. It's algorithm is composed of 5 steps:
1. Noise reduction (Gaussian filter)
2. Intensity Gradient Calculation
3. Apply Non-maximum supression to get rid of fake response
4. Apply Double threshold to determine the potential edges
5. Edge Tracking by Hysteresis by suppressing weak edges

```python
canny = cv2.Canny(img, 100, 200)
```

```python
titles = ['image', 'Laplacian', 'sobelX', 'sobelY', 'sobelXY', 'sobelXorY', 'Canny']
images = [img, laplacian, sobelX, sobelY, sobelXY, sobelXorY, canny]
```

```python
for i in range(7):
    plt.subplot(3, 3, i+1)
    plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
