# Smoothing OR Blurring Images
To remove the noise of the image. We can use diverse linear filters, as they are easy to achieve and relatively fast.

Images can be filtered with various **low pass filters** *(helps remove noise and blur the image)* and **high pass filters** *(helps in finding edges in the images)*

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./Pictures/lena.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

kernel = np.ones((5,5), np.float32)/25
```

## Linear Filters
- **Homogeneous filter:** Most simple filter, each output pixel is the mean of its kernel neighbours
- **Averaging filter:** Average of the neighbouring pixels
- **Gaussian filter:** using different weight kernel in both x and y direction. Specially designed to improve the high frequency noise from the image
- **Median filter:** Replace each pixel's value with the median of its neighbouring pixels. Great when dealing with salt and pepper noise.
- **Bilateral filter:** To disolve noise and preserve the edges.

```python
# cv2.filter2D(img, depthOfDestinationImg, kernel)
homoblur = cv2.filter2D(img, -1, kernel)

# cv2.blur(src, ksize)
avgblur = cv2.blur(img, (5,5))

# cv2.GaussianBlur(src, ksize, sigmaX)
gaussblur = cv2.GaussianBlur(img, (5,5), 0)

# cv2.medianBlur(src, odd_ksize)
medianblur = cv2.medianBlur(img, 5)

# cv2.bilateralFilter(src, diameterOfPixel, sigmaColor, sigmaSpace)
bilateralblur = cv2.bilateralFilter(img, 9, 75, 75)
```
